```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
df = pd.read_csv('resident-population-by-subzone-ethnic-group-and-sex.csv')
```

```python
by_race = df[(df['level_1'] !=  'Total')&(df['level_2'] ==  'Total')]
```

#### Within race percentage

```python
race_total = pd.DataFrame(by_race.groupby('level_1')['value'].sum()).reset_index().rename(columns={'value': 'total'})
```

```python
race_total['perc'] = race_total['total'] / race_total['total'].sum()
```

```python
by_race1 = pd.merge(by_race, race_total, how='left', on='level_1')
```

```python
by_race1['perc'] = by_race1['value'] / by_race1['total'] * 100
```

```python
by_race1.sort_values('perc', ascending=False)
```

```python
#by_race1.to_csv('by_race.csv', index=False)
```

#### Within subzone percentage

```python
subzone_total = pd.DataFrame(by_race.groupby('level_4')['value'].sum()).reset_index().rename(columns={'value': 'total'})
```

```python
by_race2 = pd.merge(by_race, subzone_total, how='left', on='level_4')
```

#### Within planning area percentage

```python
PA_perc = pd.DataFrame(by_race2.groupby(['level_3', 'level_1'], sort=True)['value'].sum())\
    .reset_index().sort_values(['level_3', 'perc'], ascending=False)
PA_perc = pd.merge(PA_perc, overall_perc, on='level_1', how='left')
PA_perc['diff'] = PA_perc['perc'] - subzone_perc['value']
```

```python
df = pd.merge(by_race1, subzone_perc.drop('perc', axis=1), on=['level_4', 'level_1'], how='left')
df = df[['level_1', 'level_4', 'perc', 'diff']]
```

```python
df = pd.pivot_table(df, columns='level_1', index='level_4', fill_value=0).reset_index()
df.columns = df.columns.to_series().str.join('_')
```

```python
df1 = df.drop('level_4_', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)

data = df_scaled.values
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6, random_state=0)
clusters = kmeans.fit_predict(data)
kmeans.cluster_centers_.shape
```

```python
plt.figure(figsize=(12,12))
plt.scatter(data[:, 0], data[:, 1], c=clusters, s=50, cmap='viridis')
```

```python
df_results = pd.concat([df, pd.Series(clusters)], axis=1)
df_results = df_results.rename(columns={'level_4_': 'subzone', '0': 'cluster'})
```
