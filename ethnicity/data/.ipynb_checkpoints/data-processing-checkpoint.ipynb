```python
import pandas as pd
```

```python
df = pd.read_csv('resident-population-by-subzone-ethnic-group-and-sex.csv')
```

```python
by_race = df[(df['level_1'] !=  'Total')&(df['level_2'] ==  'Total')]
```

#### Within race percentage

```python
race_total = pd.DataFrame(by_race.groupby('level_1')['value'].sum()).reset_index().rename(columns={'value': 'total'})
```

```python
race_total['perc'] = race_total['total'] / race_total['total'].sum()
```

```python
by_race1 = pd.merge(by_race, race_total, how='left', on='level_1')
```

```python
by_race1['perc'] = by_race1['value'] / by_race1['total'] * 100
```

```python
by_race1.sort_values('perc', ascending=False)
```

```python
#by_race1.to_csv('by_race.csv', index=False)
```

#### Within subzone percentage

```python
subzone_total = pd.DataFrame(by_race.groupby('level_4')['value'].sum()).reset_index().rename(columns={'value': 'total'})
```

```python
by_race2 = pd.merge(by_race, subzone_total, how='left', on='level_4')
```

```python
by_race2['perc'] = by_race2['value'] / by_race2['total'] * 100
```

```python
by_race2.sort_values('perc', ascending=False)
```

```python
by_race2[by_race2['level_4'] == 'Mackenzie']
```

```python
by_race2[by_race2['level_4'] == 'Western Water Catchment']
```

```python
by_race[by_race['level_4'] == 'Mackenzie']
```

```python
len(by_race['level_4'].unique())
```

```python
subzone_perc = pd.DataFrame(by_race2.groupby(['level_4', 'level_1'], sort=True)['perc'].sum())\
    .reset_index().sort_values(['level_4', 'perc'], ascending=False)
overall_perc = pd.DataFrame(by_race.groupby('level_1')['value'].sum() / by_race['value'].sum() * 100).reset_index()
subzone_perc = pd.merge(subzone_perc, overall_perc, on='level_1', how='left')
subzone_perc['diff'] = subzone_perc['perc'] - subzone_perc['value']
```

```python
#subzone_perc.to_csv('subzone_perc.csv', index=False)
```

```python
subzone_perc.sort_values('diff', ascending=False)
```

```python
subzone_perc[subzone_perc['level_1'] == 'Others'].sort_values('diff', ascending=False)
```

```python
subzone_perc[subzone_perc['level_1'] == 'Chinese'].sort_values('diff', ascending=False)
```

```python
subzone_perc[subzone_perc['level_1'] == 'Malays'].sort_values('diff', ascending=False)
```

```python
subzone_perc[subzone_perc['level_1'] == 'Indians'].sort_values('diff', ascending=False)
```

```python
contour = pd.DataFrame()
for i in [4,3,2,1]:
    t1 = subzone_perc[(subzone_perc['diff'] >= -i) & (subzone_perc['diff'] < -(i-1))]
    t1['diff_perc'] = i
    t2 = subzone_perc[(subzone_perc['diff'] > (i-1)) & (subzone_perc['diff'] <= i)]
    t2['diff_perc'] = i
    contour = pd.concat([contour, t1, t2])
```

```python
contour = pd.DataFrame()
for i in [4,3,2,1]:
    t1 = subzone_perc[(subzone_perc['diff'] >= -i) & (subzone_perc['diff'] < i)]
    t1['diff_perc'] = i
    contour = pd.concat([contour, t1])
```

```python
# Diverse subzones (high representation across all ethnic groups)
test1 = pd.DataFrame(contour[contour['diff_perc'] == 3].groupby('level_4')['diff'].count()).reset_index()
f = test1[test1['diff'] == 4]
len(f)
```
