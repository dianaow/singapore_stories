```python
import pandas as pd
```

```python
df = pd.read_csv('resident-population-by-subzone-ethnic-group-and-sex.csv')
```

```python
by_race = df[(df['level_1'] !=  'Total')&(df['level_2'] ==  'Total')]
```

#### Within race percentage

```python
race_total = pd.DataFrame(by_race.groupby('level_1')['value'].sum()).reset_index().rename(columns={'value': 'total'})
```

```python
race_total
```

```python
by_race1 = pd.merge(by_race, race_total, how='left', on='level_1')
```

```python
by_race1['perc'] = by_race1['value'] / by_race1['total'] * 100
```

```python
by_race1.sort_values('perc', ascending=False)
```

#### Within subzone percentage

```python
subzone_total = pd.DataFrame(by_race.groupby('level_4')['value'].sum()).reset_index().rename(columns={'value': 'total'})
```

```python
by_race2 = pd.merge(by_race, subzone_total, how='left', on='level_4')
```

```python
by_race2['perc'] = by_race2['value'] / by_race2['total'] * 100
```

```python
by_race2.sort_values('perc', ascending=False)
```
